### Adding/Removing Columns continue

```python
bios.head()
```

```python
bios["height_category"] = bios['height_cm'].apply(lambda x: 'Short' if x < 165 else ('Avarage' if x < 185 else 'Tall'))
```

```python
bios.head()
```

```python
bios[bios['height_category'].isin(["Tall"])]
```

```python
bios.head()
```

```python
bios.info()
```

```python
def categorize_athlete(row):
    if row['height_cm'] < 175 and row['weight_kg'] < 70:
        return 'Light Weight'
    elif row['height_cm'] < 190 or row['weight_kg'] <= 85:
        return 'Normal Weight'
    else:
        return 'Bad Weight'

bios['Category']  = bios.apply(categorize_athlete, axis=1)
```

```python
bios.head()
```
